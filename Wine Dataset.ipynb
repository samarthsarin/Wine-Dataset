# Task 3:Data Modelling:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.tree import DecisionTreeClassifier  
from sklearn import svm

from sklearn.utils import shuffle
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
 
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import accuracy_score
```

# 1. K-Nearest Neighbor Classifier:

## Parameter Selection via Grid Search for KNN:

```python
# Create list of the parameter values that should be searched
k_range = list(range(1, 21))
weight_options = ['uniform', 'distance']
p_options=[1,2]
```

```python
# create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(n_neighbors=k_range, weights=weight_options,p=p_options)
print(param_grid)
```

```python
# instantiate and fit the grid
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=10, scoring='accuracy')
grid.fit(x, y)
```

```python
grid.grid_scores_
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

## Feature Selection via Hill Climbing for KNN:

```python
new_Ind = []
cur_MaxScore = 0.0
col_num = 11
col_Ind_Random = shuffle(range(0,col_num), random_state=1)


for cur_f in range(0, col_num):
    new_Ind.append(col_Ind_Random[cur_f])
    newData1= wine.iloc[:, new_Ind]
    x_train1,x_test1, y_train1, y_test1 = train_test_split(newData1, wine.iloc[:,11], test_size=0.4, random_state=0)
    clf = KNeighborsClassifier(18,weights='distance',p=1)
    fit = clf.fit(x_train1, y_train1)
    cur_Score = clf.score(x_test1, y_test1)
    if cur_Score < cur_MaxScore:
        new_Ind.remove(col_Ind_Random[cur_f])
    else:
        cur_MaxScore = cur_Score
        print("Score with " + str(new_Ind) + " selected features: " + str(cur_Score))
```

## Displaying selected features for KNN:

```python
newData1
```

## KNN Model evaluation:

```python
y_pre = fit.predict(x_test1)  
```

```python
cm = confusion_matrix(y_test1, y_pre)
print(cm)
```

```python
print(classification_report(y_test1,y_pre))
```

```python
accuracy_score(y_test1, y_pre)
```

```python
accuracy_score(y_test1, y_pre, normalize=False)
```

# 2. Decision Tree Classifier:

## Parameter Selection via Grid Search for Decision Tree:

```python
# Create list of the parameter values that should be searched
randomstate_range = list(range(1, 21))
criterion_options = ['gini', 'entropy']
maxdepth_range = list(range(1, 21))
```

```python
# create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(criterion=criterion_options, random_state=randomstate_range,max_depth=maxdepth_range)
print(param_grid)
```

```python
# instantiate and fit the grid
grid = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=10, scoring='accuracy')
grid.fit(x, y)
```

```python
grid.grid_scores_
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

## Feature Selection for Decision Tree:

```python
new_Ind = []
cur_MaxScore = 0.0
col_num = 11
col_Ind_Random = shuffle(range(0,col_num), random_state=1)

for cur_f in range(0, col_num):
    new_Ind.append(col_Ind_Random[cur_f])
    newData2 = wine.iloc[:, new_Ind]
    x_train2, x_test2, y_train2, y_test2 = train_test_split(newData2, wine.iloc[:,11], test_size=0.4, random_state=0)
    dt_clf = DecisionTreeClassifier(criterion="entropy",random_state=5,max_depth=5)
    dt_fit = dt_clf.fit(x_train2, y_train2)
    cur_Score = dt_clf.score(x_test2, y_test2)
    if cur_Score < cur_MaxScore:
        new_Ind.remove(col_Ind_Random[cur_f])
    else:
        cur_MaxScore = cur_Score
        print ("Score with " + str(new_Ind) + " selected features: " + str(cur_Score))
```

# 3. Support Vector Machine Classifier:

## Parameter Selection via Grid Search for SVM :

```python
# Create list of the parameter values that should be searched
c_range=list(range(1,11))
kernel_options = ['linear', 'rbf']
```

```python
# create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(C=c_range,kernel=kernel_options)
print(param_grid)
```

```python
# instantiate and fit the grid
grid = GridSearchCV(svm.SVC(), param_grid, cv=10, scoring='accuracy')
grid.fit(x, y)
```

```python
grid.grid_scores_
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

## Feature Selection for SVM:

```python
new_Ind = []
cur_MaxScore = 0.0
col_num = 11
col_Ind_Random = shuffle(range(0,col_num), random_state=1)

for cur_f in range(0, col_num):
    new_Ind.append(col_Ind_Random[cur_f])
    newData3 = wine.iloc[:, new_Ind]
    x_train3, x_test3, y_train3, y_test3 = train_test_split(newData3, wine.iloc[:,11], test_size=0.4, random_state=0)
    svm_clf = svm.SVC(kernel= 'linear', C=2.0)
    svm_fit = svm_clf.fit(x_train3, y_train3)
    cur_Score = svm_clf.score(x_test3, y_test3)
    if cur_Score < cur_MaxScore:
        new_Ind.remove(col_Ind_Random[cur_f])
    else:
        cur_MaxScore = cur_Score
        print ("Score with " + str(new_Ind) + " selected features: " + str(cur_Score))
```
